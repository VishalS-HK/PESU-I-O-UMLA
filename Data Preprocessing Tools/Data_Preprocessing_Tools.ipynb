**Importing the required libraries**

```python
import pandas as pd
import matplotlib.pyplot as pt
import numpy as np
```

**Importing the Dataset**

*What is iloc() ?*

> The iloc() function in python is one of the functions defined in the PANDAS library that helps us select a specific row or column from the data set. Using the iloc() function in python, we can easily retrieve value from a row or a column using the index values. 

```python
dataset = pd.read_csv("Data.csv")
X = dataset.iloc[:, : -1].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

**Taking Care of Missing Data**

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(X[:, 1:3])
#Replacing the missing values with the mean of the column.
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

**Encoding Categorical Data**

>The below code is creating a column transformer object that will transform the first column of the X matrix into a one hot encoded matrix.

>Encoding the Independent Variable ðŸ‘‡

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])], remainder="passthrough") #[0] is the seed value
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

>Encoding the Dependent Variable ðŸ‘‡

```python

#Converting the labels into numbers.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

**Splitting the dataset into Training set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # sc is object of StandardScaler function
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```

```python
print(X_train)
```

```python
print(X_test)
```
